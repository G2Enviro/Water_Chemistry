```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load your Excel file
file_path = 'D:/G2Enviro/HDF Energy/Report/Chemistry/HDF_Chem_3.xlsx'
chemistry_input_df = pd.read_excel(file_path, sheet_name='Chemistry Input')
dwaf_1997_df = pd.read_excel(file_path, sheet_name='DWAF 1997')

# DWAF classification function
def classify_dwaf(observed_value, parameter, dwaf_df):
    """
    Classifies the observed value based on DWAF standards.
    
    Parameters:
    - observed_value: float, the observed value of the chemical constituent
    - parameter: str, name of the chemical constituent to classify
    - dwaf_df: DataFrame containing DWAF standards and classification ranges
    
    Returns:
    - classification: str, the DWAF class (Class 0, Class I, Class II, etc.)
    """
    param_row = dwaf_df[dwaf_df['Water quality substances and criteria of concern for drinking purposes.'] == parameter]
    
    if param_row.empty:
        return "Parameter not Used"
    
    param_row = param_row.iloc[0]
    
    class_ranges = {
        "Class 0": param_row["Unnamed: 2"],
        "Class I": param_row["Unnamed: 3"],
        "Class II": param_row["Unnamed: 4"],
        "Class III": param_row["Unnamed: 5"]
    }
    
    # Compare observed value to the ranges and classify
    for class_name, class_range in class_ranges.items():
        if 'to' in str(class_range):
            low, high = map(float, class_range.split(' to '))
            if low <= observed_value <= high:
                return class_name
        elif '>' in str(class_range):
            if observed_value > float(class_range.split('>')[1]):
                return class_name
        elif '<' in str(class_range):
            if observed_value < float(class_range.split('<')[1]):
                return class_name
    
    return "Unclassified"

# Custom color palette for the DWAF classifications
dwaf_palette = {
    "Class 0": "blue",
    "Class I": "green",
    "Class II": "orange",
    "Class III": "red",
    "Unclassified": "grey",
    "Parameter not Used": "black"
}

# Function to plot results for each borehole with logarithmic scaling
def plot_borehole_results(borehole_no, borehole_data):
    """
    Plots the classification of chemical constituents for a given borehole with logarithmic scaling.
    
    Parameters:
    - borehole_no: str, the borehole number (e.g., 'BOK11BH1')
    - borehole_data: DataFrame, the DataFrame containing classification results for the borehole
    """
    plt.figure(figsize=(12, 6))
    
    # Use the custom color palette for classifications
    sns.barplot(
        x='Parameter',
        y='Observed Value',
        hue='DWAF Classification',
        data=borehole_data,
        palette=dwaf_palette
    )
    
    plt.title(f'DWAF Classification for Borehole {borehole_no}')
    plt.xticks(rotation=90)  # Rotate parameter names for better readability
    plt.ylabel('Observed Value (Log Scale)')
    plt.xlabel('Chemical Constituent')
    plt.legend(title="DWAF Class")
    plt.yscale('log')  # Apply logarithmic scaling to the y-axis
    plt.tight_layout()
    
    # Save the plot to a PNG file
    plt.savefig(f"D:/G2Enviro/HDF Energy/Report/Chemistry/{borehole_no}_plot.png")
    plt.show()

# Prepare an Excel writer to save results per borehole
output_excel_file = "D:/G2Enviro/HDF Energy/Report/Chemistry/DWAF_Classification_Results_Boreholes.xlsx"
with pd.ExcelWriter(output_excel_file, engine='xlsxwriter') as writer:

    # Prepare a results DataFrame
    boreholes = chemistry_input_df.columns[1:]  # List of borehole columns
    for borehole in boreholes:
        borehole_results = pd.DataFrame(columns=["Borehole No", "Parameter", "Observed Value", "DWAF Classification"])

        # Iterate over each parameter row (after metadata like latitude, longitude)
        for index, row in chemistry_input_df.iterrows():
            if index > 4:  # Start after metadata rows
                parameter = row[0]  # The parameter name is in the first column
                if pd.isna(parameter):  # Skip empty rows
                    continue

                observed_value = row[borehole]
                
                if pd.isna(observed_value):
                    continue  # Skip missing values
                
                try:
                    observed_value = float(observed_value)  # Ensure it's a numeric value
                except ValueError:
                    continue  # Skip if the value cannot be converted to float
                
                # Classify using DWAF
                dwaf_class = classify_dwaf(observed_value, parameter, dwaf_1997_df)
                
                # Add the result to the borehole DataFrame
                new_row = pd.DataFrame({
                    "Borehole No": [borehole],
                    "Parameter": [parameter],
                    "Observed Value": [observed_value],
                    "DWAF Classification": [dwaf_class]
                })
                borehole_results = pd.concat([borehole_results, new_row], ignore_index=True)
        
        # Save the borehole-specific results to its own sheet
        borehole_results.to_excel(writer, sheet_name=borehole, index=False)
        
        # Plot and save the plot for the borehole
        plot_borehole_results(borehole, borehole_results)

print(f"Results saved to {output_excel_file}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

# Load your Excel file
file_path = 'D:/G2Enviro/HDF Energy/Report/Chemistry/HDF_Chem_3.xlsx'
chemistry_input_df = pd.read_excel(file_path, sheet_name='Chemistry Input')
sans241_df = pd.read_excel(file_path, sheet_name='SANS241')

# SANS241 classification function
def classify_sans(observed_value, substance, sans_df):
    """
    Classifies the observed value based on SANS241 standards.
    
    Parameters:
    - observed_value: float, the observed value of the chemical constituent
    - substance: str, name of the chemical constituent to classify
    - sans_df: DataFrame containing SANS241 standards and information
    
    Returns:
    - classification: str, classification result (e.g., Compliant, Exceeds Limit)
    - additional_info: dict, additional information from SANS (risk type, limit, etc.)
    """
    param_row = sans_df[sans_df['SUBSTANCE'] == substance]
    
    if param_row.empty:
        return "SUBSTANCE not found", {}

    param_row = param_row.iloc[0]
    
    # Extract the relevant fields
    standard_limit = param_row['Standard Limit']
    risk_type = param_row['Risk Type']
    water_supply_systems = param_row['Water Supply Systems']
    borehole_systems = param_row['Borehole Systems']
    additional_info = param_row['Additional Information']
    
    # Handle ranges like ' 50 to 97' and ' 10'
    try:
        if '' in str(standard_limit):
            limit_value = float(standard_limit.replace('', '').replace(',', '').strip())
            classification = "Compliant" if observed_value <= limit_value else "Exceeds Limit"
        elif '' in str(standard_limit):
            limit_value = float(standard_limit.replace('', '').replace(',', '').strip())
            classification = "Compliant" if observed_value >= limit_value else "Below Limit"
        elif 'to' in str(standard_limit):
            # Handle ranges like '50 to 97'
            range_values = [float(x.strip()) for x in standard_limit.split('to')]
            if range_values[0] <= observed_value <= range_values[1]:
                classification = "Compliant"
            elif observed_value < range_values[0]:
                classification = "Below Range"
            else:
                classification = "Above Range"
        else:
            classification = "Unclassified"
    except ValueError:
        classification = "Unclassified"

    # Store the additional information
    additional_info_dict = {
        "Risk Type": risk_type,
        "Standard Limit": standard_limit,
        "Water Supply Systems": water_supply_systems,
        "Borehole Systems": borehole_systems,
        "Additional Information": additional_info
    }
    
    return classification, additional_info_dict

# Function to plot results for each borehole with legend and logarithmic scale
def plot_borehole_results(borehole_no, df):
    """
    Plots the classification of chemical constituents for a given borehole.
    
    Parameters:
    - borehole_no: str, the borehole number
    - df: DataFrame with classification results for that borehole
    """
    plt.figure(figsize=(10, 6))
    
    # Sort data for better plot display
    df_sorted = df.sort_values(by='Observed Value')
    
    # Color mapping for classifications
    color_map = {
        "Compliant": "blue",
        "Exceeds Limit": "red",
        "Unclassified": "grey"
    }
    
    # Assign color based on classification
    colors = [color_map.get(x, 'grey') for x in df_sorted['SANS Classification']]
    
    # Create the bar plot with logarithmic scaling on the y-axis
    plt.bar(df_sorted['SUBSTANCE'], df_sorted['Observed Value'], color=colors)
    plt.xticks(rotation=90)
    plt.yscale('log')  # Apply logarithmic scale to the y-axis
    plt.title(f'SANS Classification for Borehole {borehole_no}')
    plt.ylabel('Observed Value (log scale)')
    plt.xlabel('Chemical Constituent')
    plt.tight_layout()
    
    # Add custom legend
    compliant_patch = mpatches.Patch(color='blue', label='Compliant')
    exceeds_limit_patch = mpatches.Patch(color='red', label='Exceeds Limit')
    unclassified_patch = mpatches.Patch(color='grey', label='Unclassified')
    plt.legend(handles=[compliant_patch, exceeds_limit_patch, unclassified_patch], title="Classification")
    
    # Save the plot as PNG
    plt.savefig(f"{borehole_no}_plot.png")
    plt.show()

# Prepare the Excel writer object
output_file = "D:/G2Enviro/HDF Energy/Report/Chemistry/SANS241_Boreholes_Results.xlsx"
with pd.ExcelWriter(output_file, engine='xlsxwriter') as writer:

    # Extract list of borehole columns (assuming Borehole columns start after metadata rows)
    boreholes = chemistry_input_df.columns[1:]

    # Iterate over each borehole and generate a separate sheet for each
    for borehole in boreholes:
        # Prepare a DataFrame for each borehole
        borehole_results = pd.DataFrame(columns=[
            "Borehole No", "SUBSTANCE", "Observed Value", "SANS Classification", "Risk Type", 
            "Standard Limit", "Water Supply Systems", "Borehole Systems", "Additional Information"
        ])
        
        for index, row in chemistry_input_df.iterrows():
            if index > 4:  # Start after metadata rows
                substance = row[0]  # The parameter name is in the first column
                if pd.isna(substance):  # Skip empty rows
                    continue
                
                observed_value = row[borehole]
                
                if pd.isna(observed_value):
                    continue  # Skip missing values
                
                try:
                    observed_value = float(observed_value)  # Ensure it's a numeric value
                except ValueError:
                    continue  # Skip if the value cannot be converted to float
                
                # Classify using SANS241
                sans_class, additional_info = classify_sans(observed_value, substance, sans241_df)
                
                # Append the results to the borehole-specific DataFrame
                borehole_results = pd.concat([borehole_results, pd.DataFrame({
                    "Borehole No": [borehole],
                    "SUBSTANCE": [substance],
                    "Observed Value": [observed_value],
                    "SANS Classification": [sans_class],
                    "Risk Type": [additional_info.get("Risk Type")],
                    "Standard Limit": [additional_info.get("Standard Limit")],
                    "Water Supply Systems": [additional_info.get("Water Supply Systems")],
                    "Borehole Systems": [additional_info.get("Borehole Systems")],
                    "Additional Information": [additional_info.get("Additional Information")]
                })], ignore_index=True)
        
        # Save the borehole results to a new sheet in the Excel file
        borehole_results.to_excel(writer, sheet_name=borehole, index=False)
        
        # Plot and save the plot for the borehole
        plot_borehole_results(borehole, borehole_results)

print(f"Results saved to {output_file}")
```
